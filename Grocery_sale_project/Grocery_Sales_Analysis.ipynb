```python
# # Supermart Grocery Sales Analysis & Modeling Project (with Plot Saving)
#
# **Objective:** Analyze sales data to understand performance drivers, customer behavior, and product trends, build foundational predictive models, and save all visualizations to files.

# ## 1. Setup: Import Libraries and Load Data

# %%
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import os

# Libraries for Market Basket Analysis
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

# Libraries for Time Series Forecasting
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.statespace.sarimax import SARIMAX
try:
    from prophet import Prophet
    prophet_installed = True
except ImportError:
    print("Prophet not found. Will try to install (this might take a moment).")
    import subprocess
    import sys
    try:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "prophet", "-q"])
        from prophet import Prophet
        prophet_installed = True
        print("Prophet installed successfully.")
    except Exception as e:
        print(f"Failed to install Prophet: {e}")
        prophet_installed = False

# --- Machine Learning Libraries ---
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, classification_report, confusion_matrix
from sklearn.impute import SimpleImputer

# Settings for better visualization
plt.style.use('seaborn-v0_8-whitegrid')
pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns', 50)
import warnings
warnings.filterwarnings('ignore')

# --- Directory for Plots ---
plots_dir = 'plots'
os.makedirs(plots_dir, exist_ok=True)
print(f"Plots will be saved in the directory: '{plots_dir}'")

# %%
# Load the dataset
try:
    df = pd.read_csv('Supermart Grocery Sales - Retail Analytics Dataset (1).csv')
except FileNotFoundError:
    print("ERROR: 'Supermart Grocery Sales - Retail Analytics Dataset (1).csv' not found.")
    print("Please ensure the dataset file is in the same directory as the script.")
    exit()

print("Dataset Info:")
df.info()

# ## 2. Data Cleaning & Preprocessing

# %%
# --- Date Handling ---
def parse_date(date_str):
    if pd.isna(date_str):
        return pd.NaT
    formats_to_try = ["%d-%m-%Y", "%m-%d-%Y", "%Y-%m-%d",
                      "%d/%m/%Y", "%m/%d/%Y", "%Y/%m/%d"]
    for fmt in formats_to_try:
        try:
            return pd.to_datetime(date_str, format=fmt)
        except (ValueError, TypeError):
            continue
    return pd.NaT

df['Order Date'] = df['Order Date'].apply(parse_date)
df.dropna(subset=['Order Date'], inplace=True)

# --- Derived Financial Metrics ---
df['Discount Amount'] = df['Sales'] * df['Discount']
df['Net Sales'] = df['Sales'] - df['Discount Amount']
df['COGS'] = df['Net Sales'] - df['Profit']
df['Profit Margin (%)'] = np.where(df['Net Sales'] != 0, (df['Profit'] / df['Net Sales']) * 100, 0)

# --- Feature Engineering: Time Features ---
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
df['DayOfWeek'] = df['Order Date'].dt.dayofweek # Monday=0, Sunday=6
df['WeekOfYear'] = df['Order Date'].dt.isocalendar().week.astype(int)
df['YearMonth'] = df['Order Date'].dt.to_period('M')

# --- Consistency Checks & Cleaning ---
print("\nChecking for Missing Values before extensive cleaning:")
print(df.isnull().sum())

cat_cols = ['Customer Name', 'Category', 'Sub Category', 'City', 'Region', 'State']
for col in cat_cols:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('Unknown').str.strip()

print("\nProcessed Data Head:")
print(df.head())
print("\nDescriptive Statistics:")
print(df.describe(include='all'))

# ## 3. Exploratory Data Analysis (EDA)

# ### 3.1 Overall Sales and Profit Trends
# %%
monthly_sales = df.groupby('YearMonth')['Net Sales'].sum()
monthly_profit = df.groupby('YearMonth')['Profit'].sum()
monthly_orders = df.groupby('YearMonth')['Order ID'].nunique()

fig, axes = plt.subplots(3, 1, figsize=(14, 15), sharex=True)
axes[0].plot(monthly_sales.index.to_timestamp(), monthly_sales.values, marker='o', linestyle='-')
axes[0].set_title('Monthly Net Sales Over Time')
axes[0].set_ylabel('Net Sales')
axes[0].grid(True)
axes[1].plot(monthly_profit.index.to_timestamp(), monthly_profit.values, marker='o', linestyle='-', color='green')
axes[1].set_title('Monthly Profit Over Time')
axes[1].set_ylabel('Profit')
axes[1].grid(True)
axes[2].plot(monthly_orders.index.to_timestamp(), monthly_orders.values, marker='o', linestyle='-', color='purple')
axes[2].set_title('Monthly Order Volume Over Time')
axes[2].set_ylabel('Number of Orders')
axes[2].set_xlabel('Month')
axes[2].grid(True)
plt.tight_layout()
fig.savefig(os.path.join(plots_dir, 'monthly_sales_profit_orders_trend.png'), bbox_inches='tight', dpi=300)
plt.show()

df_ts = df.set_index('Order Date')
daily_sales_ts = df_ts['Net Sales'].resample('D').sum().fillna(0)
if len(daily_sales_ts) >= 365 * 2:
    try:
        decomposition = seasonal_decompose(daily_sales_ts, model='additive', period=365)
        fig_decomp = decomposition.plot()
        fig_decomp.set_size_inches(12, 8)
        plt.suptitle('Sales Decomposition (Daily)', y=1.02)
        fig_decomp.savefig(os.path.join(plots_dir, 'sales_decomposition_daily.png'), bbox_inches='tight', dpi=300)
        plt.show()
    except Exception as e:
        print(f"\nError during seasonal decomposition: {e}. Skipping.")
else:
    print("\nNot enough data for robust seasonal decomposition with period 365. Need at least 2 full periods.")

# ### 3.2 Product Performance (Category & Sub-Category)
# %%
category_performance = df.groupby('Category').agg(
    TotalNetSales=('Net Sales', 'sum'),
    TotalProfit=('Profit', 'sum'),
    AvgProfitMargin=('Profit Margin (%)', 'mean'),
    OrderCount=('Order ID', 'nunique')
).sort_values(by='TotalNetSales', ascending=False)
print("\n--- Category Performance ---")
print(category_performance)
fig, axes = plt.subplots(1, 2, figsize=(18, 6))
sns.barplot(x=category_performance['TotalNetSales'], y=category_performance.index, ax=axes[0], palette='viridis')
axes[0].set_title('Total Net Sales by Category')
axes[0].ticklabel_format(style='plain', axis='x')
sns.barplot(x=category_performance['TotalProfit'], y=category_performance.index, ax=axes[1], palette='plasma')
axes[1].set_title('Total Profit by Category')
axes[1].ticklabel_format(style='plain', axis='x')
plt.tight_layout()
fig.savefig(os.path.join(plots_dir, 'category_sales_profit.png'), bbox_inches='tight', dpi=300)
plt.show()

subcategory_performance = df.groupby('Sub Category').agg(
    TotalNetSales=('Net Sales', 'sum'),
    TotalProfit=('Profit', 'sum'),
    AvgProfitMargin=('Profit Margin (%)', 'mean'),
    OrderCount=('Order ID', 'nunique')
).sort_values(by='TotalNetSales', ascending=False)
print("\n--- Top 15 Sub-Categories by Net Sales ---")
print(subcategory_performance.head(15))
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.barplot(x=subcategory_performance['TotalNetSales'].head(15), y=subcategory_performance.index[:15], ax=axes[0], palette='viridis')
axes[0].set_title('Top 15 Sub-Categories by Net Sales')
axes[0].ticklabel_format(style='plain', axis='x')
subcategory_performance_profit = subcategory_performance.sort_values(by='TotalProfit', ascending=False)
sns.barplot(x=subcategory_performance_profit['TotalProfit'].head(15), y=subcategory_performance_profit.index[:15], ax=axes[1], palette='plasma')
axes[1].set_title('Top 15 Sub-Categories by Profit')
axes[1].ticklabel_format(style='plain', axis='x')
plt.tight_layout()
fig.savefig(os.path.join(plots_dir, 'subcategory_top15_sales_profit.png'), bbox_inches='tight', dpi=300)
plt.show()

# ### 3.3 Geographical Performance (Region & City)
# %%
region_performance = df.groupby('Region').agg(
    TotalNetSales=('Net Sales', 'sum'),
    TotalProfit=('Profit', 'sum'),
    AvgProfitMargin=('Profit Margin (%)', 'mean'),
    OrderCount=('Order ID', 'nunique')
).sort_values(by='TotalNetSales', ascending=False)
print("\n--- Region Performance ---")
print(region_performance)
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
axes[0].pie(region_performance['TotalNetSales'], labels=region_performance.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('Set2'))
axes[0].set_title('Net Sales Distribution by Region')
sns.barplot(x=region_performance.index, y=region_performance['AvgProfitMargin'], ax=axes[1], palette='coolwarm')
axes[1].set_title('Average Profit Margin (%) by Region')
plt.tight_layout()
fig.savefig(os.path.join(plots_dir, 'region_sales_distribution_margin.png'), bbox_inches='tight', dpi=300)
plt.show()

city_performance = df.groupby('City').agg(
    TotalNetSales=('Net Sales', 'sum'),
    TotalProfit=('Profit', 'sum'),
    AvgProfitMargin=('Profit Margin (%)', 'mean'),
    OrderCount=('Order ID', 'nunique')
).sort_values(by='TotalNetSales', ascending=False)
print("\n--- Top 15 Cities by Net Sales ---")
print(city_performance.head(15))
fig, axes = plt.subplots(1, 2, figsize=(18, 8))
sns.barplot(y=city_performance.index[:15], x=city_performance['TotalNetSales'].head(15), ax=axes[0], palette='mako')
axes[0].set_title('Top 15 Cities by Net Sales')
axes[0].ticklabel_format(style='plain', axis='x')
city_performance_margin = city_performance.sort_values(by='AvgProfitMargin', ascending=False)
sns.barplot(y=city_performance_margin.index[:15], x=city_performance_margin['AvgProfitMargin'].head(15), ax=axes[1], palette='rocket')
axes[1].set_title('Top 15 Cities by Average Profit Margin (%)')
axes[1].ticklabel_format(style='plain', axis='x')
plt.tight_layout()
fig.savefig(os.path.join(plots_dir, 'city_top15_sales_margin.png'), bbox_inches='tight', dpi=300)
plt.show()

# ### 3.4 Discount Analysis
# %%
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discount', y='Profit Margin (%)', alpha=0.5)
plt.title('Profit Margin (%) vs. Discount')
plt.savefig(os.path.join(plots_dir, 'discount_vs_profit_margin_scatter.png'), bbox_inches='tight', dpi=300)
plt.show()

df['DiscountBin'] = pd.cut(df['Discount'], bins=np.arange(0, df['Discount'].max() + 0.05, 0.05), right=False)
discount_profit = df.groupby('DiscountBin', observed=False)['Profit Margin (%)'].mean().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='DiscountBin', y='Profit Margin (%)', data=discount_profit, palette='coolwarm', order=discount_profit['DiscountBin'].astype(str).sort_values())
plt.title('Average Profit Margin (%) by Discount Range')
plt.xticks(rotation=45, ha='right')
plt.savefig(os.path.join(plots_dir, 'avg_profit_margin_by_discount_range.png'), bbox_inches='tight', dpi=300)
plt.show()

discount_volume = df.groupby('DiscountBin', observed=False)['Order ID'].count().reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(x='DiscountBin', y='Order ID', data=discount_volume, palette='viridis', order=discount_volume['DiscountBin'].astype(str).sort_values())
plt.title('Order Volume (Line Items) by Discount Range')
plt.xticks(rotation=45, ha='right')
plt.savefig(os.path.join(plots_dir, 'order_volume_by_discount_range.png'), bbox_inches='tight', dpi=300)
plt.show()

# ## 4. Customer Segmentation (RFM Analysis)
# %%
snapshot_date = df['Order Date'].max() + timedelta(days=1)
rfm_data = df.groupby('Customer Name').agg(
    Recency=('Order Date', lambda x: (snapshot_date - x.max()).days),
    Frequency=('Order ID', 'nunique'),
    MonetaryValue=('Net Sales', 'sum')
)

# Robust RFM Scoring
score_cols = {'R_Score': 'Recency', 'F_Score': 'Frequency', 'M_Score': 'MonetaryValue'}
score_labels = {'R_Score': [4,3,2,1], 'F_Score': [1,2,3,4], 'M_Score': [1,2,3,4]}

for score_name, rfm_metric in score_cols.items():
    try:
        rfm_data[score_name] = pd.qcut(rfm_data[rfm_metric].rank(method='first'), q=4, labels=score_labels[score_name], duplicates='drop').astype(int)
    except ValueError as e:
        print(f"Warning during {score_name} qcut (q=4): {e}. Attempting with fewer quantiles.")
        num_quantiles = min(4, rfm_data[rfm_metric].nunique())
        if num_quantiles > 1:
            labels_dynamic = score_labels[score_name][-num_quantiles:] if score_name == 'R_Score' else score_labels[score_name][:num_quantiles]
            rfm_data[score_name] = pd.qcut(rfm_data[rfm_metric].rank(method='first'), q=num_quantiles, labels=labels_dynamic, duplicates='drop')
        else:
            rfm_data[score_name] = 1 if score_name != 'R_Score' else 4 # Assign extreme if only 1 quantile
        rfm_data[score_name] = rfm_data[score_name].astype(int)

    if rfm_data[score_name].isnull().any():
        median_val = int(np.nanmedian(rfm_data[score_name])) if rfm_data[score_name].notna().any() else (1 if score_name != 'R_Score' else 4)
        rfm_data[score_name] = rfm_data[score_name].fillna(median_val).astype(int)
    rfm_data[score_name] = np.clip(rfm_data[score_name], 1, 4)


rfm_data['RFM_Score_Combined'] = rfm_data['R_Score'].astype(str) + rfm_data['F_Score'].astype(str) + rfm_data['M_Score'].astype(str)

def assign_rfm_segment(row):
    r, f, m = row['R_Score'], row['F_Score'], row['M_Score']
    if r == 4 and f == 4 and m == 4: return 'Champions'
    if r >= 3 and f >= 3 and m >= 3: return 'Loyal Customers'
    if r == 4 and f <= 2 and m <= 2: return 'New Customers'
    if r <= 2 and f >= 3 and m >= 3: return 'Potential Loyalists (Needs Attention)'
    if r <= 2 and f <= 2 and m >= 3: return 'High Spending, Needs Attention'
    if r <= 2 and f <= 2 and m <= 2: return 'At Risk'
    if r == 1 and f <= 2 and m <= 2: return 'Hibernating/Lost'
    if r >= 3 and f <= 2 and m <= 2: return 'Promising'
    return 'Other'
rfm_data['Segment'] = rfm_data.apply(assign_rfm_segment, axis=1)

print("\nRFM Segment counts BEFORE consolidation:")
original_segment_counts = rfm_data['Segment'].value_counts()
print(original_segment_counts)
min_samples_for_stratify = 2
segments_to_consolidate = original_segment_counts[original_segment_counts < min_samples_for_stratify].index
consolidation_target_segment = 'Other - Consolidated'
if 'Other' in original_segment_counts.index and original_segment_counts['Other'] >= min_samples_for_stratify and 'Other' not in segments_to_consolidate:
    consolidation_target_segment = 'Other'
elif 'Other' in segments_to_consolidate:
     pass # Will be consolidated into 'Other - Consolidated'

if len(segments_to_consolidate) > 0:
    print(f"\nConsolidating small RFM segments into '{consolidation_target_segment}': {list(segments_to_consolidate)}")
    rfm_data['Segment'] = rfm_data['Segment'].replace(segments_to_consolidate, consolidation_target_segment)
    final_segment_counts = rfm_data['Segment'].value_counts()
    print("\nRFM Segment counts AFTER consolidation:")
    print(final_segment_counts)
    if final_segment_counts.get(consolidation_target_segment, 0) < min_samples_for_stratify and len(final_segment_counts) > 1:
        print(f"Warning: Segment '{consolidation_target_segment}' is still too small ({final_segment_counts.get(consolidation_target_segment, 0)} members).")
        if consolidation_target_segment == 'Other - Consolidated': # Only try to merge if it's our special consolidated one
            remaining_segments = final_segment_counts.drop(consolidation_target_segment, errors='ignore')
            if not remaining_segments.empty:
                largest_segment = remaining_segments.idxmax()
                print(f"Merging '{consolidation_target_segment}' into the largest remaining segment: '{largest_segment}'")
                rfm_data['Segment'] = rfm_data['Segment'].replace(consolidation_target_segment, largest_segment)
                print("\nRFM Segment counts AFTER FINAL consolidation:")
                print(rfm_data['Segment'].value_counts())

segment_analysis = rfm_data.groupby('Segment').agg(
    Count=('MonetaryValue', 'count'),
    AvgRecency=('Recency', 'mean'), AvgFrequency=('Frequency', 'mean'),
    AvgMonetary=('MonetaryValue', 'mean'), TotalMonetary=('MonetaryValue', 'sum')
).sort_values(by='TotalMonetary', ascending=False)
print("\n--- RFM Segment Analysis (Post Consolidation) ---")
print(segment_analysis)

plt.figure(figsize=(12, 7))
segment_counts_viz = rfm_data['Segment'].value_counts()
sns.barplot(x=segment_counts_viz.index, y=segment_counts_viz.values, palette='Spectral', order=segment_counts_viz.index)
plt.title('Customer Count by RFM Segment (Post Consolidation)')
plt.xticks(rotation=45, ha='right')
plt.savefig(os.path.join(plots_dir, 'rfm_segment_distribution.png'), bbox_inches='tight', dpi=300)
plt.show()

plt.figure(figsize=(12, 7))
segment_monetary_sum_viz = rfm_data.groupby('Segment')['MonetaryValue'].sum().sort_values(ascending=False)
sns.barplot(x=segment_monetary_sum_viz.index, y=segment_monetary_sum_viz.values, palette='Spectral', order=segment_monetary_sum_viz.index)
plt.title('Total Monetary Value by RFM Segment (Post Consolidation)')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45, ha='right')
plt.savefig(os.path.join(plots_dir, 'rfm_segment_monetary_value.png'), bbox_inches='tight', dpi=300)
plt.show()

# ## 5. Market Basket Analysis (Association Rules)
# %%
basket_prep_df = df.dropna(subset=['Order ID', 'Sub Category'])
basket_data = basket_prep_df.groupby('Order ID')['Sub Category'].apply(lambda x: list(set(x))).reset_index(name='Items')
if not basket_data.empty:
    te = TransactionEncoder()
    try:
        te_ary = te.fit_transform(basket_data['Items'])
        basket_df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
        min_support_threshold = 0.01
        frequent_itemsets = apriori(basket_df_encoded, min_support=min_support_threshold, use_colnames=True)
        if not frequent_itemsets.empty:
            print(f"\n--- Frequent Itemsets (min_support={min_support_threshold}) ---")
            print(frequent_itemsets.sort_values(by='support', ascending=False).head(10))
            min_confidence_threshold = 0.1
            rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence_threshold)
            if not rules.empty:
                rules_sorted = rules.sort_values(by=['lift', 'confidence'], ascending=[False, False])
                print(f"\n--- Association Rules (min_confidence={min_confidence_threshold}) ---")
                print(rules_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(15))
            else: print("No association rules found.")
        else: print("No frequent itemsets found.")
    except Exception as e: print(f"Error in Market Basket Analysis: {e}")
else: print("Basket data is empty. Skipping Market Basket Analysis.")

# ## 6. Sales Forecasting (Time Series)
# %%
sales_ts = df.groupby('Order Date')['Net Sales'].sum().resample('D').sum().fillna(0)
forecast_horizon = 90
if len(sales_ts) > forecast_horizon:
    train_data = sales_ts[:-forecast_horizon]
    test_data = sales_ts[-forecast_horizon:]
    sarima_forecast_available = False
    prophet_forecast_available = False
    try:
        print("\nFitting SARIMA model...")
        sarima_model = SARIMAX(train_data, order=(1,1,1), seasonal_order=(1,1,1,7), enforce_stationarity=False, enforce_invertibility=False)
        sarima_results = sarima_model.fit(disp=False)
        sarima_forecast_obj = sarima_results.get_forecast(steps=forecast_horizon)
        sarima_forecast = sarima_forecast_obj.predicted_mean
        sarima_conf_int = sarima_forecast_obj.conf_int(alpha=0.05)
        plt.figure(figsize=(14,7))
        plt.plot(train_data.index, train_data, label='Training Data')
        plt.plot(test_data.index, test_data, label='Actual Test Data', color='orange')
        plt.plot(sarima_forecast.index, sarima_forecast, label='SARIMA Forecast', color='green')
        plt.fill_between(sarima_conf_int.index, sarima_conf_int.iloc[:,0], sarima_conf_int.iloc[:,1], color='k', alpha=0.1, label='95% CI')
        plt.title('SARIMA Sales Forecast vs Actual'); plt.legend(); plt.grid(True)
        plt.savefig(os.path.join(plots_dir, 'forecast_sarima.png'), bbox_inches='tight', dpi=300); plt.show()
        sarima_forecast_available = True
    except Exception as e: print(f"\nError SARIMA: {e}. Skipping.")

    if prophet_installed:
        print("\nFitting Prophet model...")
        prophet_train_df = train_data.reset_index(); prophet_train_df.columns = ['ds', 'y']
        prophet_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False, interval_width=0.95)
        prophet_model.fit(prophet_train_df)
        future_dates = prophet_model.make_future_dataframe(periods=forecast_horizon, freq='D')
        prophet_forecast_df = prophet_model.predict(future_dates)
        fig_prophet1 = prophet_model.plot(prophet_forecast_df); ax_prophet1 = fig_prophet1.gca()
        ax_prophet1.plot(test_data.index, test_data.values, 'r.', label='Actual Test Data')
        ax_prophet1.set_title('Prophet Sales Forecast'); ax_prophet1.legend()
        fig_prophet1.savefig(os.path.join(plots_dir, 'forecast_prophet.png'), bbox_inches='tight', dpi=300); plt.show()
        fig_prophet2 = prophet_model.plot_components(prophet_forecast_df)
        fig_prophet2.savefig(os.path.join(plots_dir, 'forecast_prophet_components.png'), bbox_inches='tight', dpi=300); plt.show()
        prophet_forecast_test = prophet_forecast_df.set_index('ds')['yhat'][-forecast_horizon:]
        prophet_forecast_available = True
    else: print("\nProphet skipped.")

    def calculate_metrics(actual, predicted):
        actual, predicted = np.array(actual), np.array(predicted)
        mask = actual != 0
        mape = np.nan if not np.any(mask) else np.mean(np.abs((actual[mask] - predicted[mask]) / actual[mask])) * 100
        return {'MAE': mean_absolute_error(actual, predicted), 'RMSE': np.sqrt(mean_squared_error(actual, predicted)), 'MAPE (%)': mape}
    print("\n--- Forecast Evaluation Metrics (on Test Set) ---")
    if sarima_forecast_available: print("SARIMA:", calculate_metrics(test_data.values, sarima_forecast.values))
    if prophet_forecast_available: print("Prophet:", calculate_metrics(test_data.values, prophet_forecast_test.values))
else: print("\nNot enough data for train/test split in forecasting.")


# ## 7. Predictive Modeling with Machine Learning
# %%
print("\n--- Starting Predictive Modeling with Machine Learning ---")

# --- 7.1 Predicting Net Sales (Regression) ---
print("\n--- 7.1 Predicting Net Sales using RandomForestRegressor ---")
ml_df_reg_source = df.copy()
numerical_features_reg = ['Discount', 'Year', 'Month', 'Quarter', 'DayOfWeek', 'WeekOfYear']
categorical_features_reg = ['Category', 'Sub Category', 'Region', 'City']
target_reg = 'Net Sales'
ml_df_reg = ml_df_reg_source[numerical_features_reg + categorical_features_reg + [target_reg]].copy()
log_transformed_sales = False
if ml_df_reg[target_reg].skew() > 1.5 and (ml_df_reg[target_reg] > 0).all():
    print(f"Net Sales skewed ({ml_df_reg[target_reg].skew():.2f}). Log1p transforming.")
    ml_df_reg[target_reg] = np.log1p(ml_df_reg[target_reg])
    log_transformed_sales = True
ml_df_reg.dropna(inplace=True)
X_reg = ml_df_reg[numerical_features_reg + categorical_features_reg]
y_reg = ml_df_reg[target_reg]

if X_reg.empty or y_reg.empty:
    print("Not enough data for Net Sales prediction after NaN drop.")
else:
    numerical_transformer_reg = Pipeline(steps=[('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
    categorical_transformer_reg = Pipeline(steps=[('imputer', SimpleImputer(strategy='most_frequent')), ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))])
    preprocessor_reg = ColumnTransformer(transformers=[('num', numerical_transformer_reg, numerical_features_reg), ('cat', categorical_transformer_reg, categorical_features_reg)], remainder='passthrough')
    rf_regressor_pipeline = Pipeline(steps=[('preprocessor', preprocessor_reg), ('regressor', RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1))])
    X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)
    print("Training Net Sales prediction model...")
    rf_regressor_pipeline.fit(X_train_reg, y_train_reg)
    y_pred_reg_transformed = rf_regressor_pipeline.predict(X_test_reg)
    y_pred_reg = np.expm1(y_pred_reg_transformed) if log_transformed_sales else y_pred_reg_transformed
    y_test_reg_orig = np.expm1(y_test_reg) if log_transformed_sales else y_test_reg
    print(f"\nNet Sales Prediction Metrics: RMSE: {np.sqrt(mean_squared_error(y_test_reg_orig, y_pred_reg)):.2f}, R-squared: {r2_score(y_test_reg_orig, y_pred_reg):.2f}")
    try:
        ohe_feature_names = rf_regressor_pipeline.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_features_reg)
        all_feature_names_reg = numerical_features_reg + list(ohe_feature_names)
        importances_reg = rf_regressor_pipeline.named_steps['regressor'].feature_importances_
        feature_importance_df_reg = pd.DataFrame({'feature': all_feature_names_reg, 'importance': importances_reg}).sort_values(by='importance', ascending=False).head(15)
        plt.figure(figsize=(10, 8)); sns.barplot(x='importance', y='feature', data=feature_importance_df_reg, palette='crest')
        plt.title('Top 15 Feature Importances for Net Sales Prediction'); plt.tight_layout()
        plt.savefig(os.path.join(plots_dir, 'ml_sales_feature_importance.png'), bbox_inches='tight', dpi=300); plt.show()
    except Exception as e: print(f"Could not plot sales regression feature importances: {e}")

# --- 7.2 Predicting Customer RFM Segment (Classification) ---
print("\n--- 7.2 Predicting Customer RFM Segment using RandomForestClassifier ---")
if 'rfm_data' in locals() and not rfm_data.empty:
    features_cls = ['Recency', 'Frequency', 'MonetaryValue']
    target_cls = 'Segment'
    rfm_data_cls = rfm_data.dropna(subset=[target_cls] + features_cls).copy() # Use a copy
    X_cls = rfm_data_cls[features_cls]
    y_cls_str = rfm_data_cls[target_cls]

    if X_cls.empty or y_cls_str.empty:
        print("Not enough data for RFM Segment prediction after NaN drop from RFM data.")
    else:
        label_encoder = LabelEncoder()
        y_cls = label_encoder.fit_transform(y_cls_str)
        class_counts_final = np.bincount(y_cls)
        use_stratify = True
        min_required_samples = 2
        if len(class_counts_final) == 0 or np.min(class_counts_final) < min_required_samples:
            print(f"Warning: Least populated class in y_cls has {np.min(class_counts_final) if len(class_counts_final)>0 else 0} member(s). Stratification requires at least {min_required_samples}. Removing stratify.")
            use_stratify = False

        scaler_cls = StandardScaler()
        X_scaled_cls = scaler_cls.fit_transform(X_cls)

        if use_stratify:
            X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_scaled_cls, y_cls, test_size=0.2, random_state=42, stratify=y_cls)
        else:
            X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_scaled_cls, y_cls, test_size=0.2, random_state=42)

        if len(X_train_cls) == 0 or len(X_test_cls) == 0:
             print("Not enough samples for train/test split for RFM classification. Skipping model training.")
        else:
            print("Training RFM Segment prediction model...")
            rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced', n_jobs=-1)
            rf_classifier.fit(X_train_cls, y_train_cls)
            y_pred_cls = rf_classifier.predict(X_test_cls)
            accuracy_cls = accuracy_score(y_test_cls, y_pred_cls)
            print(f"\nRFM Segment Prediction Metrics: Accuracy: {accuracy_cls:.2f}")
            print("\nClassification Report:")
            present_labels_encoded = np.unique(np.concatenate((y_test_cls, y_pred_cls)))
            present_target_names = label_encoder.inverse_transform(present_labels_encoded)
            print(classification_report(y_test_cls, y_pred_cls, labels=present_labels_encoded, target_names=present_target_names, zero_division=0))

            cm = confusion_matrix(y_test_cls, y_pred_cls, labels=present_labels_encoded)
            plt.figure(figsize=(max(8, len(present_target_names)*0.8), max(6, len(present_target_names)*0.6)))
            sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=present_target_names, yticklabels=present_target_names)
            plt.title('Confusion Matrix for RFM Segment Prediction'); plt.xlabel('Predicted'); plt.ylabel('True')
            plt.xticks(rotation=45, ha='right'); plt.yticks(rotation=0); plt.tight_layout()
            plt.savefig(os.path.join(plots_dir, 'ml_rfm_segment_confusion_matrix.png'), bbox_inches='tight', dpi=300); plt.show()

            importances_cls = rf_classifier.feature_importances_
            feature_importance_df_cls = pd.DataFrame({'feature': features_cls, 'importance': importances_cls}).sort_values(by='importance', ascending=False)
            plt.figure(figsize=(8,5)); sns.barplot(x='importance', y='feature', data=feature_importance_df_cls, palette='viridis')
            plt.title('Feature Importances for RFM Segment Prediction'); plt.tight_layout()
            plt.savefig(os.path.join(plots_dir, 'ml_rfm_segment_feature_importance.png'), bbox_inches='tight', dpi=300); plt.show()
else:
    print("RFM data not available/empty. Skipping RFM Segment prediction.")

# ## 8. Conclusions & Recommendations
# %%
print(f"\nGenerated plots in '{plots_dir}':")
try:
    print(os.listdir(plots_dir))
except FileNotFoundError:
    print(f"Directory '{plots_dir}' not created.")
print("\n--- End of Analysis and Modeling ---")
```
